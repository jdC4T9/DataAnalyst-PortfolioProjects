```python
def api_runner():
    global df
    
    url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
    parameters = {
      'start':'1',
      'limit':'15',
      'convert':'USD'
    }
    headers = {
      'Accepts': 'application/json',
      'X-CMC_PRO_API_KEY': '07a15077-090b-4ecb-8e2b-72b7eb1fe35a',
    }

    session = Session()
    session.headers.update(headers)

    try:
      response = session.get(url, params=parameters)
      data = json.loads(response.text)
      print(data)
    except (ConnectionError, Timeout, TooManyRedirects) as e:
      print(e)
    
    
    #This normalizes the data and makes it all pretty in a variable named 'df' aka dataframe
    df = pd.json_normalize(data['data'])
    df['timestamp'] = pd.to_datetime('now')
    df
    
    
    # to create a csv and append data to it
    if not os.path.isfile(r'C:\Users\Jd\AlexTheAnalyst-PythonProjects-Acer\Project9_Python-Automating-API-Pull.csv'):
        df.to_csv(r'C:\Users\Jd\AlexTheAnalyst-PythonProjects-Acer\Project9_Python-Automating-API-Pull.csv', header='column_names')
    else:
        df.to_csv(r'C:\Users\Jd\AlexTheAnalyst-PythonProjects-Acer\Project9_Python-Automating-API-Pull.csv', mode='a', header=False)
```

```python
# importing the necessary libraries 
import os 
from time import time
from time import sleep

# creating a for loop to automatically pull data from the crypto website: coinmarketcap.com
# range is set to 333 to maximize the free 333 credits for API Key usage each day
for i in range(333):
    api_runner()
    print('API Runner completed')
    sleep(60) #sleep for 1 minute
exit()
```

```python
# looking at the coin trends over time
# data gathered in the .csv file are limited to 4 API pulls since we set it to pull data every 1 minute
# and we only let that for loop run for 4mins, hence, the data gathered after 4 API pulls

coinTrends = df.groupby('name', sort=False)[['quote.USD.percent_change_1h','quote.USD.percent_change_24h','quote.USD.percent_change_7d','quote.USD.percent_change_30d','quote.USD.percent_change_60d','quote.USD.percent_change_90d']].mean()
coinTrends
```

```python
# stacking the columns per name
coinTrends_stacked = coinTrends.stack()
coinTrends_stacked
```

```python
# converting the coinTrends_stacked that is of type 'Series' back t being a 'DataFrame'

coinTrends_stacked_df = coinTrends_stacked.to_frame(name='values')
coinTrends_stacked_df
```

```python
# because of how the DataFrame is structured above we need to set an index
# i don't want to pass a column as an index for this dataframe
# so I'm going to create a range and pass that as the dataframe
# NOTE: you can make this more dynamic, but I'm just going to hard code it

index = pd.Index(range(90))


# set the above DataFrame index object as the index
# using set_index() function
coinTrends_stacked_df_indexed = coinTrends_stacked_df.reset_index()
coinTrends_stacked_df_indexed


# if it only has the index and values try doing reset_index like "coinTrends_stacked_df.reset_index()"
```
